## Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
# merging datasets on title

movies = movies.merge(credits,on='title')
```

## Selecting Columns

#### In a nutshell we are creating a content base recomendar system for that we need to create tags so we will select columns from the dataset of our need.
#### We are selecting genres(kind of movies), id(will help in fetching posters in end),  keywords(info about movie), title, cast, crew(will focus on director)

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

## Preprocessing

```python
##checking null values

movies.isnull().sum()
```

```python
# since missing values are not that much so droping them is not an issue
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

#### Dealing with genres

```python
movies.iloc[0].genres

# our data is in dictionary form and we need to convert it into a list 
# i.e ['Action','Adventure','Fantasy']
```

```python
# our list is in string format so converting it into a list we use ast function which we will do our work
import ast
```

```python
def convert(obj):
    list1 = []
    for i in ast.literal_eval(obj):
        list1.append(i['name'])
    return list1
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

#### Dealing with keywords

```python
movies['keywords'] = movies['keywords'].apply(convert) 
```

```python
movies.head()
```

#### Dealing with cast
##### as we want only top three actors of the movie 

```python
def convert2(obj):
    list1 = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            list1.append(i['name'])
            counter = counter + 1
        else:
            break
    return list1
```

```python
movies['cast'].apply(convert2)
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

#### Dealing with crew

###### As we want only director name

```python
def fetch_director(obj):
    list1 = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list1.append(i['name'])
            break
    return list1
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

#### Dealing with overview

```python
## overview is in string we will convert into list so we concatenate other lists with it

movies['overview'][0]
```

```python
# split function return the string in the form of list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

#### Removing spaces among words in columns 

###### We are doing this so our recomender system don't get confuse

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

#### Transformation on Columns 

```python
## Creating a tag column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
## removing unnecessary columns

new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
### Convert tags list data into string

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'][0]
```

```python
## converting into lower caps

new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#converting tags into vectors
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

##### here we can see that accept, accepted and accepts are same so we need to make them with only one word i.e. accept so for this we are using nltk function to cope with this issue.

```python
cv.get_feature_names()
```

```python
import nltk
```

##### applying stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)  # converting back to string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

##### Now we need to calculate the distance between two vectors of the movies, we can't use eculidean distance because it fails in higher dimensions so for this we use cosine distance for getting the similarity between the vectors of the movies

#### importing cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#calculating the similarity between vectors
similarity = cosine_similarity(vector)
```

```python
# similarity of every movie with another movie and diagonally they are similar
similarity[0]
```

##### Here we are getting the top 5 similar movies

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

##### As we can see that our model is giving us similar movies to Batman Begins. This dataset has 5000 movies data so we can easily check each and every movie here and get a suggestion on it.
