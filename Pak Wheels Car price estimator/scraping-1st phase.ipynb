```python
final=pd.DataFrame()
for j in range(1,208):
  webpage=requests.get('https://www.pakwheels.com/used-cars/search/-/mk_suzuki/md_cultus/?page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_='col-md-9 grid-style')
  name=[]
  price=[]
  city=[]
  year=[]
  mileage=[]
  feul=[]
  engine=[]
  driving=[]
  

  for i in company:

    try:
       name.append(i.find('h3').text.strip())
    except:
       name.append(np.nan)
       

    try:
       price.append(i.find('div',class_='price-details generic-dark-grey').text.strip())
    except:
       price.append(np.nan)
      
   
    try:

      city.append(i.find('ul' , class_='list-unstyled search-vehicle-info fs13').text.strip())
    except:
      city.append(np.nan)
    
    
    try:
      
      list1 = i.find('ul',class_='list-unstyled search-vehicle-info-2 fs13')
      items = list1.find_all('li')
    
      lists = [ [x] for x in items ]
      year.append(lists[0])
      mileage.append(lists[1])
      feul.append(lists[2])
      engine.append(lists[3])
      driving.append(lists[4])
        
    except:
        #year.append(np.nan)
        #mileage.append(np.nan)
        #feul.append(np.nan)
        #engine.append(np.nan)
        #driving.append(np.nan)
        pass
    


    
  df=pd.DataFrame({'name':name,
         'price':price,
       'city':city,
        'year':year,
       'mileage':mileage,
       'feul':feul,
      'engine':engine,
      'driving':driving,
    })
 
  final=final.append(df,ignore_index=True)   
    
  
```
