## Cleaning the features

```python
for r in (("cc", ""), ("<li>", ""),("</li>",""),("\n",""),("   ",""),("[",""),("]","")):
    df4['driving'] = df4['driving'].str.replace(*r,regex=True)
    
```

```python
parts = df4['price'].str.split('\.')
part0 = parts.str[0].str.extract('(\d+)').fillna('0')
part1 = parts.str[1].str.extract('(\d+)').fillna('0')
df4['price'] = part0 + "." + part1
```

```python
df4['mileage']= df4['mileage'].str.replace(r'\D', '')
```
